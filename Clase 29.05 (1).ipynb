## Interpolaci√≥n por Splines

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
```

```python
A= np.array([
    [1, 0, 1, 0],
    [0,1,0,-2],
    [1, -1, 3, 0],
    [0, 0, 1, 1]
])
b= np.array([1, 2, 0, 0])
sol = np.linalg.solve(A,b)
print(sol)
```

```python
x = sp.symbols('x')

S0 = 2+ 0.75 * (x-1)+0.25* (x-1)**3
display(sp.expand(S0))

S1= 3 + 1.5* (x-2) + 0.75 * (x-2)**2 - 0.25*(x-2)**3
display(sp.expand(S1))
```

```python
S0_N = sp.lambdify(x, S0)
S1_n = sp.lambdify(x, S1)
```

```python
x= np.array([1, 2, 3])
y = np.array([2, 3, 5])

x_interp0= np.linspace(x[0], x[1], 100)
x_interp1= np.linspace(x[1], x[2], 100)

plt.plot(x, y, color='hotpink', label='Datos', marker='o')
plt.plot(x_interp0, S0_N(x_interp0), label= 'S_0(x)')
plt.plot(x_interp1, S1_n(x_interp1), label= 'S_1(x)')
plt.grid()
plt.legend()
plt.show()
```
